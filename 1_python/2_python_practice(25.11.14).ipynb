```python
# 문제15 - set으로 중복 제거 + 정렬
students = ["Kim", "Lee", "Kim", "Park", "Choi", "Lee", "Lee"]
stu = set(["Kim", "Lee", "Kim", "Park", "Choi", "Lee", "Lee"])
print(stu)
```

```python
# 문제16 - 집합 연산 (합/교/차)
user1 = {"SF", "Action", "Drama"}
user2 = {"Drama", "Romance", "Action"}
print(f"공통 장르: {user1 & user2}")
print(f"서로 다른 장르: {user1 ^ user2}")
print(f"모든 장르: {user1 | user2}")
```

```python
# 문제17 - 부분집합 판정
mine = {"SQL", "PYTHON", "LINUX"}
required = {"PYTHON", "SQL"}
print(f"지원가능여부: {mine.issuperset(required)}")
```
