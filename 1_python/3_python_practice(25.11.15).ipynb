```python
#문제2. 문자열 중간 통째로 뒤집기
text = "pythonstudy"
text1 = text[0:3]
text2 = text[3:][::-1]
print(text1 + text2)
```

```python
#문제3. 리스트에서 양 끝 요소만 추출
items = ["A", "B", "C", "D", "E"]
items1 = [items[0], items[-1]]
print(items1)
```

```python
#문제4. 리스트에서 홀수 인덱스만 추출
nums = [5, 10, 15, 20, 25, 30, 35]
print(nums[1::2])
```

```python
#문제5. 슬라이싱으로 문자열 반 갈라서 교체
word = "abcdefgh"
word1 = word[0:4]
word2 = word[4:][::-1]
print(word1 + word2)
```

```python
#문제6. 리스트에서 3칸마다 하나씩 추출
data = [1, 2, 3, 4, 5, 6, 7, 8, 9]
print(data[::3])
```

```python
#문제7. 문자열의 짝수 인덱스만 출력
msg = "HELLOWORLD"
print(msg[::2])
```

```python
#문제8. 리스트 뒷 부분만 두 번 반복
nums = [10, 20, 30, 40, 50]
nums1 = nums[2:] * 2
print(nums1)
```

```python
#문제9. 날짜 문자열 분리 후 형식 바꾸기
data = "2025-11-15".split("-")
print(f"{data[0]}년 {data[1]}월 {data[2]}일")
```

```python
#문제10. 리스트 단순 병합 후 중간 만 출력
a = [1, 2, 3]
b = [4, 5, 6, 7]
c = a + b
print(c[2:5])
```

```python
#문제11. 튜플 값 교체 (불변성 활용)
info = ("user", 30, "korea")
my_info = ("Doohee",) + info[1:]
print(my_info)
```

```python
#문제12. 튜플 언패킹 후 문자열 조합
data = ("Python", "3.13", "Language")
program, version, language = data
print(f"{program}은 {version} 버전의 {language} 입니다.")
```

```python
#문제13. 튜플에서 뒤에서 3개만 추출
t = (10, 20, 30, 40, 50, 60)
print(t[3:])
```

```python
#문제14. 두 튜플 합쳐서 짝수만 걸러내기
a = (1, 2, 3)
b = (4, 5, 6)
c = a + b
print(c[1::2])
```

```python
#문제15. set으로 중복제거 + 사전 순 정렬
names = ["Kim", "Lee", "Kim", "Park", "Choi", "Lee"]
names1 = sorted(set(names))
print(names1)
```

```python
#문제16. set 연산 3종
A = {"Apple", "Banana", "Kiwi"}
B = {"Banana", "Melon", "Apple"}
print(f"교집합: {A & B}")
print(f"합집합: {A | B}")
print(f"차집합: {A - B}")
```

```python
#문제17. 부분집합 판정
skill_have = {"SQL", "Python"}
skill_need = {"Python"}
print(skill_need.issubset(skill_have))
```

```python
#문제18. 문자열 -> set 변환
msg = "SUCCESS"
print(set(msg))
```

```python
#문제19. 대칭 차집합
x = {"A", "B", "C", "D"}
y = {"C", "D", "E"}
print(f"대칭 차집합: {x ^ y}")
```

```python
#문제20. 튜플에서 타입별로 분리하기
data = ("A", 1, "B", 2, "C", 3, "A", 1)
data1 = data[::2]
data2 = data[1::2]
data3 = set(data1)
data4 = tuple(sorted(set(data2)))
print(f"""문자열: {data3}
숫자열: {data4}""")

```
